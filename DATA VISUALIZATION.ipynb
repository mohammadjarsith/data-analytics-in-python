# scatter plot

```python
movies=['mi1','fast&furious','avengers','viduthalai','speed','pollathavan','vada chennai']
percentage=[41,45,67,78,56,87,92]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(movies,percentage)
```

```python
plt.plot(movies,percentage,linestyle='solid')#solid,dashed,dotted
```

```python
plt.bar(movies,percentage)
```

```python
plt.pie(percentage,labels=movies)
```

# working with inbulit datasets

```python
sns.get_dataset_names()
```

```python
tip=sns.load_dataset('tips')
```

```python
tip
```

```python
tip.tail()
```

# relationship plot

```python
sns.relplot(x='total_bill',y='tip',data=tip,col='time',hue='day',style='day',kind='scatter')
```

# categorical box plot

```python
sns.catplot(x='day',y='total_bill',kind='box',data=tip)
```

# regression plot

```python
sns.regplot(x='total_bill',y='tip',data=tip)
```
